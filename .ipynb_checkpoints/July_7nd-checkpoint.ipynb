## Existing Problems

### assembleλmatrix() being bottleneck for memory
This function seems to be the bottle neck of the whole code. For large problem, I encounter load error when using backslash inside this function where the left hand-side is the cholesky factorization results for sparse arrays.<br>
For problems that I can solve without this issue, this part is expensive in time also.


### How to measure memory accurately in Julia
The benchmark seems to give a memory allocations for all intermediate allocations that will be recycled. But I want to measure the maximum memory allocation for a piece of code because the size of the system that is solvable given system memory is restricted by this. I got this result when the system memory that I required is only 128 GBs, and from slurm job information, the maximum memory allocation is only around 18 GB

There are some basic simple multithreading examples in Julia. Since multithreading with @spawn was introduced after Julia 1.3, before that it was tested @threads. I was trying to find more multithreading examples in a complex setting but failed. I tried to re-write some part of the SBP code to include multithreading techniques, but got errors from Julia that I couldn't really understand well. It would be great to have some more materials explaing new multithreading mechanism in Julia with complex problem settings
